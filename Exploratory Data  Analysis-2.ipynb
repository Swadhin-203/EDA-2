 ## insights and Observation
 Based on the boxplot of flight prices, there seem to be some potential outliers in the dataset. Outliers are data points that lie outside the range of typical values and may skew the distribution of data.

Outliers can have a significant impact on statistical analysis, as they may affect measures of central tendency such as the mean and median, as well as measures of variability such as the standard deviation and range. For example, if we were to calculate the mean price of flights including the outliers, the resulting value may not accurately represent the typical price of flights, as the outliers may pull the mean in one direction or another. Similarly, the standard deviation of the prices may be larger if the outliers are included, which may indicate more variability than is actually present in the typical prices.

Therefore, it is important to carefully consider the presence of outliers and decide whether to exclude them or investigate them further in our analysis. One approach is to use a statistical test, such as the Tukey method, to identify and remove outliers. Another approach is to examine the data and determine whether the outliers are legitimate values or errors in data entry, and to decide whether to keep or remove them based on this assessment.

The missing values may impact the analysis in several ways, such as:

Incomplete information: If there are missing values in a critical field, it may not be possible to draw meaningful conclusions or insights from the dataset. For instance, if there are missing values in the price or rating columns, it may be difficult to determine the price and quality of the apps accurately.

Skewed results: If there are missing values in a particular category or field, it may skew the results of any analysis that uses that field. For instance, if there are missing values in the app category column, it may skew the analysis of how ratings vary by category.

Biased results: If there is a significant number of missing values in a particular field, it may lead to biased results that do not accurately represent the population. For instance, if there are missing values in the app size column, it may skew the analysis of how app size impacts ratings, leading to incorrect conclusions.

Therefore, it is important to identify and handle missing values appropriately to ensure the accuracy and reliability of any analysis or insights derived from the dataset.

```python
# Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
# the relationship.
import pandas as pd
import numpy as np

# Load the Google Playstore dataset
google_playstore = pd.read_csv("gh.csv")
google_playstore ['Size']=google_playstore ['Size'].str.replace('M','000')
google_playstore ['Size']=google_playstore ['Size'].str.replace('k','')
google_playstore ['Size']=google_playstore ['Size'].replace('Varies with device',np.nan)
google_playstore ['Size']=google_playstore ['Size'].replace('1,000+',np.nan)

google_playstore ['Size']=google_playstore ['Size'].astype(float)


# Remove missing values from the dataset
google_playstore.dropna(subset=['Size', 'Rating'], inplace=True)

import matplotlib.pyplot as plt

# Create a scatter plot to visualize the relationship between the size of an app and its rating
google_playstore.plot(kind='scatter', x='Size', y='Rating', alpha=0.5)

# Set the x-axis label
plt.xlabel('Size (in KB)')

# Set the y-axis label
plt.ylabel('Rating')

```

```python
# Q15. A company wants to launch a new app on the Google Playstore and has asked you to analyze the
# Google Playstore dataset to identify the most popular app categories. How would you approach this
# task, and what features would you analyze to make recommendations to the company?
# Ans:
import pandas as pd

# Load the Google Playstore dataset
google_playstore = pd.read_csv("gh.csv")

# Remove missing values from the dataset
google_playstore.dropna(subset=['Installs'], inplace=True)
# Define a custom function to convert the 'Installs' column to a numeric data type
def convert_installs(x):
    x = x.replace(',', '')
    x = x.replace('+', '')
    return int(x)
google_playstore["Installs"]=google_playstore["Installs"].replace("Free","10")
# Apply the custom function to the 'Installs' column
google_playstore['Installs'] = google_playstore['Installs'].apply(convert_installs)
# Count the number of apps in each category
category_counts = google_playstore['Category'].value_counts()

import matplotlib.pyplot as plt

# Create a bar chart of the number of apps in each category
plt.figure(figsize=(12,6))
plt.bar(category_counts.index, category_counts.values)
plt.xticks(rotation=90)
plt.xlabel('App Category')
plt.ylabel('Number of Apps')
plt.title('Distribution of App Categories in the Google Playstore')
plt.show()
```

```python
# Calculate the average number of installs for each category
installs_by_category = google_playstore.groupby('Category')['Installs'].mean()
# Create a bar chart of the average number of installs for each category
plt.figure(figsize=(12,6))
plt.bar(installs_by_category.index, installs_by_category.values)
plt.xticks(rotation=90)
plt.xlabel('App Category')
plt.ylabel('Average Number of Installs')
plt.title('Popularity of App Categories in the Google Playstore')
plt.show()
```
